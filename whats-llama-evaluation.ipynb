```python
import numpy as np
import matplotlib.pyplot as plt

# Data
data ={
    'Q1': {'oli': 53, 'gpt4-prompt-data': 35, 'Llama-2-7b-chat-hf-whatsllama': 68, 'Llama-2-german-whatsllama': 51, 'gpt4-no-data': 28, 'Llama-2-7b-german-assistant-v3': 10},
    'Q2': {'oli': 92, 'gpt4-prompt-data': 40, 'Llama-2-7b-chat-hf-whatsllama': 22, 'Llama-2-german-whatsllama': 22, 'gpt4-no-data': 25, 'Llama-2-7b-german-assistant-v3': 38},
    'Q3': {'oli': 83, 'gpt4-prompt-data': 44, 'Llama-2-7b-chat-hf-whatsllama': 43, 'Llama-2-german-whatsllama': 39, 'gpt4-no-data': 21, 'Llama-2-7b-german-assistant-v3': 17},
    'Q4': {'oli': 35, 'gpt4-prompt-data': 68, 'Llama-2-7b-chat-hf-whatsllama': 47, 'Llama-2-german-whatsllama': 50, 'gpt4-no-data': 22, 'Llama-2-7b-german-assistant-v3': 26}
}

# Redefining sorted_keys based on the updated data
sorted_keys = data['Q1'].keys()  # As provided by you earlier

# Generating the stacked bar chart as per the provided code snippet
bottom_values = np.zeros(len(sorted_keys))

fig, ax = plt.subplots(figsize=(12, 8))

colors = ['blue', 'green', 'red', 'purple', 'cyan', 'orange']
for question, color in zip(data.values(), colors):
    scores = [question.get(key, 0) for key in sorted_keys]
    ax.bar(sorted_keys, scores, bottom=bottom_values, color=color)
    bottom_values += scores

# Adding labels, title, and legend
ax.set_ylabel('Scores')
ax.set_title('Summary of Model Performance')
ax.set_xticklabels(sorted_keys, rotation=45, ha='right')
ax.legend(data.keys())
plt.tight_layout()
plt.show()
```
